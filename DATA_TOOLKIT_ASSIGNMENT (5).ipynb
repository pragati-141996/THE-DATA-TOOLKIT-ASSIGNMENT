1. What is NumPy, and why is it widely used in Python ?
-  NumPy is a Python library that provides:
-  Multidimensional array objects (called ndarray)
-  A collection of mathematical functions to operate on these arrays
-  Tools for linear algebra, Fourier transforms, random number generation, and more
- It is widely used due to the following reasons:
-  Efficient Array Operations : NumPy arrays are faster and more memory-efficient than native Python lists, especially for large datasets and numerical calculations.
-  Vectorization : You can perform operations on entire arrays without explicit loops (e.g., a + b where a and b are arrays). This makes code both faster and cleaner.
-  Interoperability : It integrates well with C/C++, Fortran, and other Python libraries like Pandas, Matplotlib, Scikit-learn, and TensorFlow.
-  Standard in Data Science & AI : NumPy is the foundation for most data analysis and machine learning libraries in Python — if you're working with data, you're probably using NumPy.
-  Broadcasting : It lets you perform operations on arrays of different shapes — a super powerful feature for mathematical modeling.

2. How does broadcasting work in NumPy ?

-  Broadcasting in NumPy is a technique that allows arrays of different shapes and sizes to be used together in mathematical operations. NumPy automatically expands the smaller array so it matches the shape of the larger one — without copying data.
-  When performing operations between arrays:NumPy compares their shapes starting from the last dimension.If the sizes are the same, or one of them is 1, NumPy can stretch or broadcast it.If dimensions are not compatible, it raises an error.

6. What are the differences between NumPy arrays and Python lists ?

-  The differences between NumPy arrays and Python lists are as follows:
-  Data Type Consistency:Python Lists can store elements of different data types (e.g., integers, strings, floats, etc.).Whereas,NumPy Arrays can only store elements of the same data type, which makes them more efficient for numerical operations.
-  Speed and Performance:Python Lists are slower when performing mathematical operations.Whereas,NumPy Arrays are faster and optimized for numerical computations.
-  Memory Efficiency:Python Lists use more memory because they store type information for each element.Whereas,NumPy Arrays are more memory efficient, as they store data in a compact form.
-  Mathematical Operations:Python Lists do not support element-wise operations directly. For example, adding two lists joins them together.
NumPy Arrays support element-wise operations (e.g., addition, subtraction, multiplication) easily.
-  Multidimensional Capability:Python Lists can be nested to form multi-dimensional lists, but working with them becomes complicated.Whereas,NumPy Arrays naturally support multi-dimensional arrays and make it easy to work with matrices or tensors.
-  Functionality and Tools:Python Lists offer basic functionality for general-purpose programming.Whereas,NumPy Arrays provide a wide range of mathematical functions (e.g., mean, sum, standard deviation) and are widely used in scientific computing.

7. What is a heatmap, and when should it be used ?

-  A heatmap is a type of data visualization that uses colors to represent the values of data in a matrix (a table of numbers). Each cell in the matrix is colored based on its value — for example, lighter colors might show lower values and darker colors might show higher values.
-  When to Use a Heatmap:
-  To show correlations between variables : For example, in a dataset, you can use a heatmap to show how strongly one feature is related to another.
-  To analyze a large table of values quickly : Instead of reading numbers, the viewer can understand the data by just looking at colors.
-  To detect patterns or anomalies : Heatmaps make it easier to notice which areas have higher or lower values at a glance.
-  To visualize frequency or intensity : For example, in website analytics, a heatmap can show where users click most often.

8. What does the term “vectorized operation” mean in NumPy ?
-  In NumPy, a vectorized operation refers to performing operations on entire arrays (vectors, matrices, etc.) at once, without using explicit loops like for or while.Instead of iterating element by element (as you'd do in pure Python), NumPy applies the operation to all elements at once internally using optimized C code. This makes the operations much faster and more efficient.

15. How does NumPy handle multidimensional arrays ?

-  A shape attribute that defines its dimensions (e.g., (3, 4) for a 2D array).

18. What are some advanced features of NumPy ?
-  The advanced features of NumPy are :
-  Broadcasting: Allows operations on arrays of different shapes by automatically adjusting their dimensions to match, improving performance without extra memory usage.
-  Vectorization: Enables element-wise operations on arrays without explicit loops, speeding up execution by leveraging low-level C implementations.
-  Advanced Indexing and Slicing: Supports boolean arrays, fancy indexing, and conditional slicing for flexible and efficient data access.
-  Memory Views and Strides: Allows manipulation of data without copying it, saving memory and improving performance, especially with large datasets.
-  Linear Algebra: Provides functions for matrix operations, eigenvalues, determinants, and other linear algebra tasks (e.g., np.dot(), np.linalg.inv()).
-  Random Number Generation: Offers tools for generating random numbers from various distributions, useful for simulations and statistical analysis.
-  ufuncs (Universal Functions): Vectorized functions that enable fast element-wise operations on arrays, such as mathematical and logical operations.
-  Structured Arrays: Stores heterogeneous data types in an array-like structure, useful for datasets with mixed types (e.g., CSV files).
-  Memory Efficiency with np.memmap: Allows large arrays to be stored on disk and accessed as if they were in memory, useful for working with large datasets that don't fit in RAM.

19. How does Pandas simplify time series analysis ?
-  Pandas simplify time series analysis through following ways :
-  Datetime Indexing: Allows efficient indexing and slicing of time series data with DatetimeIndex, enabling time-based selections and operations.
-  Resampling: Supports changing the frequency of time series data (e.g., from daily to monthly) using resample() for aggregation or up/downsampling.
-  Date Offsets: Enables easy shifting of data by specific time periods (e.g., lags or leads) with shift() and date_offset functionalities.
-  Time Zone Handling: Simplifies working with time zones, including localizing and converting time zones with tz_localize() and tz_convert().
-  Rolling and Moving Window Functions: Provides simple methods for calculating rolling statistics like moving averages using rolling().
-  Shifting and Lagging: Allows creating lagged columns to compare past values using shift() for time-based analysis.
-  Handling Missing Data: Facilitates dealing with missing time series data through methods like forward filling (ffill()), backward filling, or interpolation.
-  Time Series Decomposition: Supports time series decomposition for identifying trends, seasonality, and residuals, often used with external libraries like statsmodels.

21. Why is NumPy’s array slicing faster than Python’s list slicing ?
-  NumPy’s array slicing is faster than Python’s list slicing because:
-  Contiguous memory: NumPy arrays are stored in continuous blocks, enabling faster access.
-  Fixed data types: Allows for efficient, low-level operations.
-  Returns views: Slicing creates a view (not a copy), saving time and memory.
-  C-level optimization: NumPy operations are implemented in fast, low-level C code.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
dataf=sns.load_dataset("iris")
corr=dataf.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5)
plt.title("CORRELATION MATRIX")
plt.show()
```
